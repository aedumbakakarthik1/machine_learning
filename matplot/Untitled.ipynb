```python
over_50 =df[ df["age"]>50]
over_50.head()
```

```python
#pyplot method
over_50.plot(kind="scatter",x="age",y="chol",c="target");
```

```python
#oo method mixed with pyplot method
fix,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter"
            ,x="age",
            y="chol",
            c="target",
            ax=ax);
ax.set_xlim([50,67])
```

```python
# oo method from scratch
fx, ax= plt.subplots(figsize=(10,6))

#ploting the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# customize plotting 
ax.set(title="heart disease and cholesterol  levels",
      xlabel="age",
      ylabel="Cholesterol")
# adding the legned
ax.legend(*scatter.legend_elements(),title="Target");

# adding a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle="--",c="green");

#adding  a  vertical line
ax.axvline(over_50["age"].mean(),linestyle="--",c="red");
```

```python
#subplot of chol , age, thalach
# for chol and age 
fig, (ax1,ax2)= plt.subplots(nrows=2,
                            ncols=1,figsize=(10,6),sharex=True)
#scatter the plot
scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"]
                   )
ax1.set(title="heart disease and cholesterol  levels",
#       xlabel="age",
      ylabel="Cholesterol");
# adding the legned
ax1.legend(*scatter.legend_elements(),title="Target");

# adding a horizontal line
ax1.axhline(over_50["chol"].mean(),linestyle="--",c="green");



# for age and thalach
scatter = ax2.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])


ax2.set(title="heart disease  with age and thalach",
       xlabel="Age",
       ylabel="thalach");

# adding legned
ax2.legend(*scatter.legend_elements(),title="Target");

# adding horizontal line
ax2.axhline(over_50["thalach"].mean(),linestyle="--");


# add a title for fig

fig.suptitle("Heart Disease Analysis", fontsize=16,fontweight="bold");
```
